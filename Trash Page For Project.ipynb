```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import optimize
```

```python
Tides = np.loadtxt('ASTR19_S22_group_project_data.txt', dtype = "str")  
T_Height_ft= np.array(Tides[:,2], dtype=float)
T_Time_hr_min= np.array(Tides[:,1], dtype=str)
T_Days= np.array(Tides[:,0], dtype= float)
y=T_Height_ft
x=T_Days

f=plt.figure(figsize=(25,7))
plt.errorbar(x,y,fmt='o')

def f_line(x,A,B,C,D):
    return A * np.sin(B * x) * np.cos(C * D)

params, params_cov = optimize.curve_fit(f_line,x,y)
A_fit= params[0]
B_fit= params[1]
C_fit= params[2]
D_fit= params[3]

print(A_fit,B_fit,C_fit,D_fit)

y_fit=  A_fit * np.sin(B_fit * x) * np.cos(C_fit * D_fit)

plt.plot(x,y_fit)
```
