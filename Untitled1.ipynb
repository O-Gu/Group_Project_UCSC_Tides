```python
#https://github.com/O-Gu/Group_Project_UCSC_Tides
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from scipy.optimize import curve_fit
import statistics as stats
with open('ASTR19_S22_group_project_data.txt') as f:
    lines = f.readlines()
    day = [line.split()[1] for line in lines]
    tide = [line.split()[2] for line in lines]
```

```python
data = np.loadtxt('ASTR19_S22_group_project_data.txt', dtype = "str")

day, time, tide = [],[],[]

for i in data:
        day = np.append(day, int(i[0]))
        time = np.append(time, i[1])
        tide = np.append(tide, float(i[2]))
def func(x, a, b, c):
    return a * np.exp(-b * x) + c
popt, pcov = curve_fit(func, day, tide)
plt.plot(day, func(day, *popt), 'r-', label='fit')

plt.plot(time, tide)
plt.scatter(time, tide)
```

```python
resi = sns.residplot(x=day, y=tide, data=data)
#plt.hist(resi) 
b, a = np.polyfit(day, tide, deg=1)
plt.show()
```
